```python
import pandas as pd
import numpy as np
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pyodbc as pyo
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
from selenium.webdriver.chrome.options import Options
```

```python
# Kéo dữ liệu nhà thầu được phê duyệt
driver = webdriver.Chrome(executable_path="./chromedriver.exe")
#option = webdriver.ChromeOptions()
#driver = webdriver.Chrome(options = option)
driver.get("https://muasamcong.mpi.gov.vn/web/guest/approved-contractors-list")
WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.XPATH, '//*[@id="search-approved-contractor"]/div/div/main/div/div/div/div[2]/div[1]/div[1]')))
objective_page = driver.find_element(By.XPATH,'//*[@id="search-approved-contractor"]/div/div/main/div/div/div/div[2]/div[2]/div[1]/div/ul/li[6]')
m = 1
script = "arguments[0].innerText = '" + str(m) + "'"
driver.execute_script(script,objective_page)
time.sleep(3)
objective_page.click()
time.sleep(15)
i=m
final_result = []
connection = pyo.connect(driver='{ODBC Driver 17 for SQL Server}'
                        ,host='LT32046SME035'
                        ,database='SME'
                        ,trusted_connection='yes')
cursor = connection.cursor()
while i in range(0,100000):
    soup = BeautifulSoup(driver.page_source,'html.parser')
    body_items = soup.find_all('div',class_='content__body__item')
    for body_item in body_items:
        body_item_above = body_item.find('div',class_='content__body__item__heading')
        Ma_dinh_danh = body_item_above.find('h4').text.strip()
        Trang_thai_HD = body_item_above.find('span').text.strip()
        body_item_below = body_item.find('div',class_='d-flex flex-column align-items-start gap-16')
        Ten_DN = body_item_below.find('h2').text.strip()
        body_item_below_titles = body_item_below.find_all('p',class_='content__body__item__body__title m-0')
        pre_1 = []
        for body_item_below_title in body_item_below_titles:
            body_item_below_title_nextsibling = body_item_below_title.find_next_sibling('p')
            pre_1.append(body_item_below_title_nextsibling.text.strip())
        Ma_so_thue = pre_1[0]
        Dia_chi = pre_1[1]
        Ngay_cap_ma = pre_1[2]
        Ngay_cap_ma = datetime.strptime(Ngay_cap_ma, '%d/%m/%Y')
        Chu_the = 'Nhà đầu tư được phê duyệt'
        now = datetime.now()
        dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
        Ngay_keo_du_lieu = datetime.strptime(dt_string, "%d/%m/%Y %H:%M:%S")
        sql = "Insert into SME..TAIMX_SRC_DANH_SACH_DINH_DANH ([Ma_dinh_danh],[Trang_thai_HD],[Ma_so_thue],[Dia_chi],[Ngay_cap_ma],[Chu_the],[Ngay_keo_du_lieu],[Page_num],[Ten_DN]) Values(?,?,?,?,?,?,?,?,?)"
        cursor.execute(sql,(Ma_dinh_danh,Trang_thai_HD,Ma_so_thue,Dia_chi,Ngay_cap_ma,Chu_the,Ngay_keo_du_lieu,i,Ten_DN))
        connection.commit()
        print('Inserted successfully for MST:',Ma_so_thue)
    print('Written successfully for page number',i)
    print('-----------------')
    i=i+1
    objective_page = driver.find_element(By.XPATH,'//*[@id="search-approved-contractor"]/div/div/main/div/div/div/div[2]/div[2]/div[1]/div/ul/li[6]')
    driver.execute_script("arguments[0].setAttribute('value',arguments[1])",objective_page,i)
    objective_page.click()
    time.sleep(20)
```
